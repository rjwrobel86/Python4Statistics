```python
#Find all files in folder 1 that are not in folder 2

import os

def get_all_files(folder):
    file_set = set()
    for root, _, files in os.walk(folder):
        for file in files:
            relative_path = os.path.relpath(os.path.join(root, file), folder)
            file_set.add(relative_path)
    return file_set

def find_missing_files(folder1, folder2):
    files_in_folder1 = get_all_files(folder1)
    files_in_folder2 = get_all_files(folder2)
    
    missing_files = files_in_folder1 - files_in_folder2
    return missing_files

#Example usage
folder1 = "/Users/robertwrobel/Library/Mobile Documents/com~apple~CloudDocs/Compiler /HME/New Folder With Items"
folder2 = "/Volumes/Crucial X9/Backup"

missing_files = find_missing_files(folder1, folder2)
for file in missing_files:
    print(file)

print(f"{len(missing_files)} missing files remaining")
```
