Pytrends - Unofficial Google Trends API

```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
used_kw_list = ["used cars for sale", "used trucks for sale", "used suvs for sale", "used car dealerships"]
preowned_kw_list = ["preowned cars for sale", "preowned trucks for sale", "preowned suvs for sale", "preowned car dealerships"]
car_site_list = ["cars.com", "cargurus", "autotrader", "kelly blue book", "carfax"]
weber_list = ["weber chevrolet", "weber granite city chevy", "george weber chevy"]
```

```python
pytrends.build_payload(used_kw_list, cat=0, timeframe='today 5-y', geo='US-MO-609', gprop='')
df = pytrends.interest_over_time()
#df['demand'] = df['cars.com'] + df['cargurus'] + df['autotrader'] + df['kelly blue book'] + df['carfax']
```

```python
df['used_demand_terms'] = df['used cars for sale'] + df['used trucks for sale'] + df['used suvs for sale'] + df['used car dealerships']
#df['preowned_demand_terms'] = df['preowned cars for sale'] + df['preowned trucks for sale'] + df['preowned suvs for sale'] + df['preowned car dealerships']

#df['used_demand_sites'] = df['cars.com'] + df['cargurus'] + df['autotrader'] + df['kelly blue book'] + df['carfax']
sites = df['used_demand_terms']
```
