# Linear Interpolation for Missing Values

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/auto48.csv', na_values='.')
df.head()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index, df['TERMCBAUTO48NS'])
plt.title('New Auto Loan Interest Rates - 48 Months')
plt.xlabel('Date')
plt.ylabel('Interest Rate')
plt.grid(True)
plt.show()
```

```python
df.interpolate(method='linear', inplace=True)
#df.to_csv('auto48interpolated.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.index, df['TERMCBAUTO48NS'])
plt.title('New Auto Loan Interest Rates - 48 Months')
plt.xlabel('Date')
plt.ylabel('Interest Rate')
plt.grid(True)
plt.show()
```
