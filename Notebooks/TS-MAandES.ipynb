# Moving Averages and Exponential Smoothing 

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
#Generate example time series data
np.random.seed(1663)
dates = pd.date_range(start='1990-01-01', periods=100, freq='M') #Make a column of dates
dates = pd.to_datetime(dates) #Tell Pandas to treat the column as a datetime object
values = np.cumsum(np.random.randn(100))
df = pd.DataFrame({'Date': dates, 'Value': values})
df.set_index('Date', inplace=True) 

#Plot the time series data
plt.figure(figsize=(10, 4))
plt.plot(df.index, df['Value'])
plt.title('Monthly Profit Margin')
plt.xlabel('Date')
plt.ylabel('% Profit')
plt.grid(True)
plt.show()
```

```python
#Moving / Rolling Average 
#Calculate a 3 period moving average
window_size = 3
df['Moving_Avg_3'] = df['Value'].rolling(window=window_size).mean()

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(df.index, df['Moving_Avg_3'], color='red', label='Three Period Moving Average')
plt.title('Original Data vs Moving Average')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Moving / Rolling Average 
#Calculate a 12 period moving average
window_size = 12
df['Moving_Avg_12'] = df['Value'].rolling(window=window_size).mean()

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(df.index, df['Moving_Avg_12'], color='red', label='12 Period Moving Average')
plt.title('Original Data vs Moving Average')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Calculate 3, 6, and 12 Period Moving Averages
window_sizes = [3,6,12]
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Value'], label='Original Data')

for i in window_sizes:
    df['Moving_Avg_' + str(i)] = df['Value'].rolling(window=i).mean()
    plt.plot(df.index, df['Moving_Avg_' + str(i)], label=f'{i} Period Moving Average')

plt.title('Original Data vs Moving Averages')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Exponential Smoothing (Weighted Average) w/ Pandas
alpha = 0.1  #Smoothing factor
df['Exponential_Smoothed'] = df['Value'].ewm(alpha=alpha, adjust=False).mean()

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(df.index, df['Exponential_Smoothed'], color='red', label=f'Exponentially Smoothed, Alpha={alpha}')
plt.title('Original Data vs Exponentially Smoothed')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
