```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import grangercausalitytests
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n = 100
x = np.random.normal(size=n)
y1 = np.zeros(n)
y2 = np.zeros(n)

#Create a causal relationship where y1 depends on past x and y2 doesn't
for t in range(2, n):
    y1[t] = 0.5 * y1[t-1] + 0.3 * x[t-1] + np.random.normal()
    y2[t] = 0.5 * y2[t-1] + np.random.normal()


plt.figure(figsize=(10,5))
plt.plot(x, label='x (Potential Cause)')
plt.plot(y1, label='y (Potential Effect)')
plt.plot(y2, label='y2')
plt.legend()
plt.show()
```
