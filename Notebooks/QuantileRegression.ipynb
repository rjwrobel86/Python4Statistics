# Quantile Regresson

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/SalesData.csv')
df.head()
```

```python
ols_model = smf.ols('TotalSales ~ TotalInventory + AverageTemperature + UsedCPI + q2 + q3 + q4', df)
ols_model_result = ols_model.fit()
ols_model_result.summary()
```

```python
#25th Percentile
quantile_model_25 = smf.quantreg('TotalSales ~ TotalInventory + AverageTemperature + UsedCPI + q2 + q3 + q4', df)
quantile_result_25 = quantile_model_25.fit(q=0.25)
quantile_result_25.summary()
```

```python
#Median Quantile
quantile_model = smf.quantreg('TotalSales ~ TotalInventory + AverageTemperature + UsedCPI + q2 + q3 + q4', df)
median_quantile_result = quantile_model.fit(q=0.5)
median_quantile_result.summary()
```

```python
#75th Percentile
quantile_model_75 = smf.quantreg('TotalSales ~ TotalInventory + AverageTemperature + UsedCPI + q2 + q3 + q4', df)
quantile_result_75 = quantile_model_75.fit(q=0.75)
quantile_result_75.summary()
```

```python
model = smf.quantreg('TotalSales ~ TotalInventory + AverageTemperature + UsedCPI + q2 + q3 + q4', df)
list_of_quantiles = [0.10,0.25,0.75,0.99]
params_list = []

for i in list_of_quantiles:
    result = model.fit(q=i)
    params_list.append(result.params)
    
params_list
```
