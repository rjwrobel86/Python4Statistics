# Basic Graphics with Matplotlib

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate fake data
#Set a random seed for reproducibility
np.random.seed(7134)
x = np.linspace(0, 10, 100) #Using linspace gives even distanced values
y = np.sin(x) + np.random.normal(0, 0.2, 100) #Using normal gives normally distributed values
z = [np.random.normal(0, std, 100) for std in range(1, 5)] #Using normal again, but with a list comprehension
categories = ['A', 'B', 'C', 'D', 'E'] #Categorical variables
categories2 = ['A2','B2','C2','D2','E2']
values = np.random.randint(1, 10, len(categories)) #Using randint gives random integer values
```

```python
#Line plot
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Bar Chart
plt.figure(figsize=(8, 4))
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(8, 4))
plt.hist(y, bins=10)
plt.title('Histogram')
plt.xlabel('y')
plt.ylabel('Frequency')
plt.show()
```

```python
#Box & Whisker Plot
plt.figure(figsize=(8, 4))
plt.boxplot(z)
plt.title('Box and Whisker Plot')
plt.xlabel('Data')
plt.ylabel('Value')
plt.show()
```

```python
#Pie chart
plt.figure(figsize=(8, 4))
plt.pie(values, labels=categories, autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Multiple Graphics Using Subplots
plt.figure(figsize=(10, 4))

#Bar plot
plt.subplot(131)
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

#Pie chart
plt.subplot(132)
plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Pie Chart')

#Box and Whisker Plot
plt.subplot(133)
plt.boxplot(z)
plt.title('Box and Whisker Plot')
plt.xlabel('Data')
plt.ylabel('Value')

plt.show()
```

```python
#Contingency Tables

import pandas as pd

data = {
    'Category_A': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'A'],
    'Category_B': ['X', 'Y', 'X', 'Y', 'Y', 'X', 'X', 'Y']
}

df = pd.DataFrame(data)

#Create a contingency table using the crosstab function
contingency_table = pd.crosstab(df['Category_A'], df['Category_B'])
print(contingency_table)

#Create a contingency table with row percentages
contingency_table_row_percentages = pd.crosstab(df['Category_A'], df['Category_B'], normalize='index')
print(contingency_table_row_percentages)

#Create a contingency table with column percentages
contingency_table_column_percentages = pd.crosstab(df['Category_A'], df['Category_B'], normalize='columns')
print(contingency_table_column_percentages)

#Create a contingency table with row and column totals
contingency_table_with_totals = pd.crosstab(df['Category_A'], df['Category_B'], margins=True, margins_name="Total")
print(contingency_table_with_totals)

#Create a contingency table with percentages of total count
contingency_table = pd.crosstab(df['Category_A'], df['Category_B'])
#Calculate the grand total
grand_total = contingency_table.values.sum()
contingency_table_percentages = (contingency_table / grand_total) * 100

print(contingency_table_percentages)

```

```python
import random

colors = ['Red', 'Blue', 'Green', 'Yellow', 'Orange']

#Generate random counts between 1 and 20 for each color
counts = [random.randint(1, 20) for _ in colors] #List comprehension

#Create a DataFrame with the colors and counts
data = {'Color': colors, 'Count': counts}
df = pd.DataFrame(data)

#Create a summary table with totals
summary_table_totals = df.groupby('Color')['Count'].sum().reset_index() #Reset index changes the formatting

#Calculate the total count
total_count = summary_table_totals['Count'].sum()

#Create another summary table with percentages
summary_table_percentages = summary_table_totals.copy()
summary_table_percentages['Percentage'] = (summary_table_percentages['Count'] / total_count) * 100


print(summary_table_totals)
print(summary_table_percentages)
```
