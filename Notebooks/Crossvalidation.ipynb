```python
#Crossvalidation 
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('../data/murder.csv')
df.head(50)
```

```python
X = df[['executions','south','unemployment','capital_punishment_legal','income_per_capita']]
y = df['murderper100k']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=1663)

mse_values = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model = sm.OLS(y_train, X_train).fit()
    print(model.params)
    print("\n")
    
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

mse_values = np.array(mse_values)
print(mse_values)

print(f'Mean MSE: {mse_values.mean()}')
print(f'Standard Deviation of MSE: {mse_values.std()}')
```
