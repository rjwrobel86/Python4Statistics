# Analysis of Covariance (ANCOVA)

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Salary': [60000, 65000, 70000, 72000, 75000, 50000, 52000, 58000, 61000, 63000],
    'Major': ['Engineering', 'Engineering', 'Engineering', 'Business', 'Business',
              'Arts', 'Arts', 'Arts', 'Business', 'Engineering'],
    'Age': [25, 27, 29, 24, 26, 30, 33, 35, 28, 22]  
}

df = pd.DataFrame(data)
```

```python
#ANCOVA model - ANOVA controlling for age
model = smf.ols('Salary ~ Major + Age', data=df).fit()
ancova_table = sm.stats.anova_lm(model, typ=2)
print(ancova_table)
```

```python
#Residual plot
residuals = model.resid
sns.histplot(residuals)
plt.title("Residual Distribution")
plt.show()

#QQ plot for normality check
sm.qqplot(residuals, line='45', fit=True)
plt.show()
```
