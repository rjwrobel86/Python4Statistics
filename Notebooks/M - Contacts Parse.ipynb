```python
import pandas as pd
import re

df = pd.read_csv('/Users/robertwrobel/Downloads/messages.csv')
```

```python
def clean_number(input_str):
    if not isinstance(input_str, str):  # Handle non-string inputs
        input_str = str(input_str) if pd.notnull(input_str) else ''
    # Remove all non-numeric characters
    cleaned = re.sub(r'[^\d\s]', '', input_str)
    # Remove whitespace
    cleaned = cleaned.replace(' ', '')
    # Remove leading 1s
    cleaned = re.sub(r'^1+', '', cleaned)
    return cleaned


# Apply the function to the column
df['Number'] = df['Number'].apply(clean_number)

# Display the DataFrame
print(df)
```

```python
df2 = pd.read_csv('/Users/robertwrobel/Downloads/contacts 2.csv')
df2['Mobile'] = df2['Mobile'].apply(clean_number)
df2['Home'] = df2['Home'].apply(clean_number)
df2['Home'] = df2['Home'].apply(clean_number)
df2['Name'] = df2['First Name'] + ' ' + df2['Last Name']
df2.head()
```

```python
# Step 1: Reshape df2 to create a lookup table
lookup = df2.melt(id_vars='Name', var_name='type', value_name='Number').dropna()

# Step 2: Merge df with the lookup table
df = df.merge(lookup, on='Number', how='left')

# Step 3: Add "who" column
df['Who'] = df['Name'] + ' (' + df['type'] + ')'
```

```python
df = df[df['Number'].astype(str).str.len() > 6]

# Display the final DataFrame
df.to_csv('/Users/robertwrobel/Downloads/test2.csv')
```
